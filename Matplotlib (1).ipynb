#Matplotlib
Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPY stack. It was introduced by 
John Hunter in the year 2002.
One of the greatest benefits of visualization is that it allows us visual access to huge amount of data in easily digestible visuals.

Installation of Matplotlib
If you have Python and PIP already installed on system, then installation of Matplotlib is very easy.
Install it using the command :
C:\Users\Your Name>pip install matplotlib

Pyplot 
Most of the Matplotlib utilies lies under the pyplot submodule , and are usually imported under alias:
import matplotlib.pyplot as plt

Types of graph which can be made using Matplotlib :

1. Linear Plot
2. Scatter Plot
3. Bar Plot
4. Stem
5. Hist Plot
6. Box Plot
7. Pie Plot
8. Fill_Between_Plot

```python
x = [5,8,3,9,10]
y = [10,7,3,4,9]
plt.plot(x,y, marker = 'o') #this will plot the dots on the graph
plt.show()
```

```python
#To get different marker for both the axis
x= [1,5,8,2]
y = [10,6,9,8]

plt.plot(x,y)
plt.plot(y, marker = '<')#Marker for y-axis
plt.plot(x, marker = 'o')#Marker for x-axis
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([1,2,3,4])

plt.bar(x,y, color= "#4CAF50", width = 0.2)#This will give width 
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([1,9,3,7])

plt.barh(x,y, color= "#4CAF50", height = 0.1)#Here we used barh function to get the bar horizontally with the height size mention in function
plt.show()
```


#Scatter Plot


```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.scatter(x,y)
plt.show()
```

```python
x = [4,5,6,7,2,6]
y = [2,3,4,8,10,8]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("number")
c = ['red','green','orange','yellow','blue','brown']

plt.scatter(x,y, color = c)
plt.show()
```

```python
x = [4,5,6,7,2,6]
y = [2,3,4,8,10,8]

plt.xlabel('Month')
plt.ylabel('Number')
plt.title('Scatter Plot')

c=['red','orange','blue','green','yellow','brown']

plt.scatter(x,y,color = c, s= 300) #here s will give the size to the dots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r'IHH.jpg'

#opening the image using PIL
image = Image.open(fname).convert('L')
#mapping image to gray scale
plt.imshow(image, cmap = 'coolwarm')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r'bird.jpeg'
plt.title('Bird')
plt.xlabel('length')
plt.ylabel('breadth')

image = Image.open(fname).convert('L')
plt.imshow(image, cmap = 'rainbow_r' )#This will change the color of image
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r'bird.jpeg'
plt.title('Bird')
plt.xlabel('length')
plt.ylabel('breadth')

image = Image.open(fname).convert('L')
plt.imshow(image, cmap = 'rainbow_r' )#This will change the color of image
plt.grid()
plt.show()
```

#Matplotlib Pie Chart

```python
x=[10,20,30,40]
y=['english','hindi','maths','science']

plt.pie(x, labels = y)
plt.show()
```

```python
x=[20,45,78,90]
y=['english','hindi','maths','science']
plt.pie(x,labels=y)
plt.legend()#This will give the following changes
plt.show()
```
